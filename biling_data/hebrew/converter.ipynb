```python
abbreviations = ["ptg","wat","grf","as","fd","bp","sh","fr","aw","dw","pr","pp","qw","qa","hv","a","v","t","c","o","p"]
```

```python
def get_word(x):
    for a in abbreviations:
        if a in x[:3]:
            return x.replace(a,"")
        else:
            return x
def get_abb(x):
    for a in abbreviations:
        if a in x[:3]:
            return a 
```
