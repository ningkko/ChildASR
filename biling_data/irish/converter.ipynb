```python
import re
def clean(x):
    # for irish
    x = x.replace("e#","é").replace("a#","á").replace("u#",'ú').replace("o#","ó").replace('i#','í').replace("@",'(')
    # for @@ -> ()
    content = re.compile( "\@(.*)\@" ).search(x)
    if content:
        content = "@"+content.group(1)+"@"
        x.replace(content,'('+content+')')

ddd = pd.DataFrame(df_sum.columns.to_list(), columns = ["words"])
ddd["words"] = ddd["words"].apply(lambda x: clean(x))
ddd.to_csv("index.csv",index=False)
```
